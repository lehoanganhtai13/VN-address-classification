```python
from address_segmentation_ import addressSegmentation
import time
import json
import re
```

```python
dataset = open('./../dataset/vn_address_dataset.txt', encoding="utf8").readlines()
segmenter = addressSegmentation(dataset)
```

```python
test_file = open('./../tests/test_cases/public.json')
test_data = json.load(test_file)
```

```python
str(type(test_data[0]))
```

```python
pass_count = 0
for data in test_data:
    address = data['text']
    start = time.time()
    result = segmenter.segment(address)
    stop = time.time()

    pass_value = False
    if result['province'] == data['result']['province'] and result['district'] == data['result']['district'] and result['ward'] == data['result']['ward']:
        pass_value = True
    # print('Pass value:', pass_value)
    pass_time = True if ((stop - start) < 0.1) else False
    # print('Pass time:', pass_value)
    if pass_value == True and pass_time == True:
        pass_count += 1
    else:   
        print(address)
        print('result:',result)
        print('answer',data['result'])
        print('')

print('Total pass:', pass_count)
```
